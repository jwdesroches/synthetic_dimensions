```python
N = 2
M = 2
sites = []

for n in range(N):
    for m in range(M):
        sites.append(f"{n},{m}>")
        
state_counter = 0
for i in [0,1]:
    for j in [0,1]:
        for k in [0,1]:
            for l in [0,1]:
                print(f"state {state_counter}")
                print(f"|{i},{sites[0]} ⊗ |{j},{sites[1]} ⊗ |{k},{sites[2]} ⊗ |{l},{sites[3]}")
                state_counter += 1
                print()
```

```python
def generate_sites(N, M):
    sites = []
    for n in range(N):
        for m in range(M):
            sites.append(f"{n},{m}>")
    return sites

def generate_states(N, M):
    import itertools
    
    sites = generate_sites(N, M)
    num_sites = N * M
    states = list(itertools.product([0, 1], repeat=num_sites))
    
    state_counter = 0
    all_states = []
    for state in states:
        print(f"state {state_counter}")
        state_repr = " ⊗ ".join(f"|{bit},{site}" for bit, site in zip(state, sites))
        print(state_repr)
        state_counter += 1
        print()
        
        all_states.append(state_repr)
    
    return state_counter, all_states

N = 3
M = 2
total_states, all_states = generate_states(N, M)
```

```python
for index_1, i in enumerate(all_states):
    for index_2, j in enumerate(all_states):
        print(f"Matrix Element: H_({index_1},{index_2})")
        print("left state:", i)
        print("right state:", j)
        print()
```
