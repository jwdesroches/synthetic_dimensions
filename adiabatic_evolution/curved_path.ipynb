```python
import sys
import os
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), os.pardir)))
from synth_dim_model import *

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap
import pandas as pd

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("energy_gap_over_V_neg_V_resolution=1000_25x25.csv", index_col='J/V_ratios')
energy_gap_grid = df.to_numpy()

J_V_ratios = np.linspace(-25,25,1000)
mu_V_ratios = np.linspace(0,25,1000)
mu_V_ratio_grid, J_V_ratio_grid = np.meshgrid(mu_V_ratios, J_V_ratios)
```

```python
plt.figure()
contour_values = np.arange(7.7,8.7, 0.1)
plt.contour(J_V_ratio_grid, mu_V_ratio_grid, energy_gap_grid, shading = "auto", cmap = "plasma", levels = contour_values)
plt.colorbar(label="$\Delta E/ |V|$")
plt.xlabel("$J/|V|$")
plt.ylabel("$\mu/|V|$")
plt.tight_layout()
```
