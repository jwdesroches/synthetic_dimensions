```python
import sys
import os
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), os.pardir)))
from synth_dim_model import *

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap
import pandas as pd

import warnings
warnings.filterwarnings("ignore")
```

```python
resolution = 1000
df = pd.read_csv(f"energy_gap_over_V_resolution={resolution}.csv").astype("float")
energy_gap_over_V_grid = df.to_numpy().T

min_V_J_ratio = -3
max_V_J_ratio = 3
max_mu_V_ratio = 3

mu_V_ratios = np.linspace(0, max_mu_V_ratio, resolution)
J_V_ratios = np.linspace(min_V_J_ratio, max_V_J_ratio, resolution)
```

```python
plt.figure(figsize = (8,6))
plt.pcolormesh(J_V_ratios, mu_V_ratios, energy_gap_over_V_grid, shading = "auto", cmap = "plasma")

plt.title(f"Energy Gap: $N={2}$, $M={5}$, $V<0$") # specify depending on data parameters
plt.colorbar(label="$\Delta E / |V|$")
plt.xlabel("$J/|V|$")
plt.ylabel("$\mu/|V|$")
```
