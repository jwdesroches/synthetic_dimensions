```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("paraparticle_asp_results.csv")

Ns = [2, 3, 4, 5]
colors = ["r", "g", "b", "m"]

fig, ax = plt.subplots()
for i, N in enumerate(Ns):
    mask = (df["N"] == N) & (df["J_V_final"] == 0) & (df["M"] == 3) & (df["V"] == -1)
    subset = df[mask].sort_values(by="T")
    x_data = subset["T"]
    y_data = subset["ground_state_overlap"]
    ax.plot(x_data, y_data, '--', color=colors[i])
    ax.plot(x_data, y_data, '.', markersize = 7.5, color=colors[i], label=f"$N={N}$")


ax.set_xlabel("$T$")
ax.set_ylabel(r"$\langle \psi(T) | P_D | \psi(T) \rangle$")
ax.set_title("ASP Results - Paraparticles: $M=3$, $(J/V)_f = 0$, $V<0$")
ax.set_ylim(-0.05, 1.05)
ax.grid()
ax.legend()
plt.show()
```

```python
df = pd.read_csv("gas_results.csv")

Ns = [2, 3, 4, 5]
Ms = [2, 3, 4, 5]
colors = ["r", "g", "b", "m"]

for M in Ms:
    fig, ax = plt.subplots()
    for i, N in enumerate(Ns):
        mask = (df["N"] == N) & (df["J_V_final"] == 2.5) & (df["M"] == M) & (df["V"] == 1)
        subset = df[mask].sort_values(by="T") 
        x_data = subset["T"]
        y_data = subset["ground_state_overlap"]
        ax.plot(x_data, y_data, '--', color=colors[i])
        ax.plot(x_data, y_data, '.', markersize = 7.5, color=colors[i], label=f"$N={N}$")


    ax.set_xlabel("$T$")
    ax.set_ylabel(r"$\langle \psi(T) | P_D | \psi(T) \rangle$")
    ax.set_title(f"ASP Results - Symmetric String: $M={M}$, $(J/V)_f = 2.5$, $V<0$")
    ax.set_ylim(-0.05, 1.05)
    ax.grid()
    ax.legend()
    plt.show()
```
