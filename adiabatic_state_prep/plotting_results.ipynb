```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("combined_asp_results.csv")
```

```python

Ns = [2, 3, 4, 5]
colors = ["r", "g", "b", "m"]

fig, ax = plt.subplots(figsize = (8,3))
for i, N in enumerate(Ns):
    mask = (df["N"] == N) & (df["J_V_final"] == 0) & (df["M"] == 3) & (df["V"] == -1)
    subset = df[mask].sort_values(by="T")
    x_data = subset["T"]
    y_data = subset["ground_state_overlap"]
    ax.plot(x_data, y_data, '--', color=colors[i])
    ax.plot(x_data, y_data, '.', markersize = 7.5, color=colors[i], label=f"$N={N}$")


ax.set_xlabel("Total Evolution Time $T$ $[\hbar/V]$")
ax.set_ylabel(r"Fidelity $\mathcal{F} = \langle \psi(T) | P_D | \psi(T) \rangle$")
ax.set_title("ASP - Free Paraparticle Phase: $M=3$, $(J/V)_f = 0$, $V<0$")
ax.set_ylim(-0.05, 1.05)
ax.grid()
ax.legend()
plt.show()
```

```python

Ns = [2, 3, 4, 5]
Ms = [2, 3, 4, 5]
colors = ["r", "g", "b", "m"]

for N in Ns:
    fig, ax = plt.subplots(figsize = (8,3))
    for i, M in enumerate(Ms):
        mask = (df["N"] == N) & (df["J_V_final"] == 2.5) & (df["M"] == M) & (df["V"] == 1)
        subset = df[mask].sort_values(by="T") 
        x_data = subset["T"]
        y_data = subset["ground_state_overlap"]
        ax.plot(x_data, y_data, '--', color=colors[i])
        ax.plot(x_data, y_data, '.', markersize = 7.5, color=colors[i], label=f"$M={M}$")


        ax.set_xlabel("Total Evolution Time $T$ $[\hbar/V]$")
    ax.set_ylabel(r"Fidelity $\mathcal{F} = \langle \psi(T) | P_D | \psi(T) \rangle$")
    ax.set_title(f"ASP - Gas Phase: $N={N}$, $(J/V)_f = 2.5$, $V>0$")
    ax.set_ylim(-0.05, 1.05)
    ax.set_xlim(0,5)
    ax.grid()
    ax.legend()
    plt.show()
```
