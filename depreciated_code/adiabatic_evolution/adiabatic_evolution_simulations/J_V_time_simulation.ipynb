```python
N = 2
M = 5
init_J = 0.01
init_V = -0.01
init_mu = 10
final_mu = 0
dt = 0.1

# Define t_totals and J/V ratios
t_totals = np.linspace(1, 25, 25)
V_J_ratios = np.linspace(-2, 2, 25)  # Adjust range as needed

# Initialize array to store ground state overlaps
ground_state_overlaps = np.zeros((len(t_totals), len(V_J_ratios)))

# Loop over t_totals and J/V ratios
for i, t_total in enumerate(t_totals):
    for j, ratio in enumerate(V_J_ratios):
        final_J = 1
        final_V = ratio
        
        adiabatic_energies, adiabatic_diff, adiabatic_wavefunctions, adiabatic_overlaps, true_energies, energy_gaps, times = \
            calculate_adiabatic_properties(N, M, init_mu, init_J, init_V, final_J, final_V, final_mu, t_total, dt, initial_state=None)
        
        ground_state_overlaps[i, j] = adiabatic_overlaps[-1, 0]

# Plot the colormap
plt.figure(figsize=(8, 6))
plt.imshow(ground_state_overlaps, extent=[V_J_ratios.min(), V_J_ratios.max(), t_totals.min(), t_totals.max()],
           origin='lower', aspect='auto', cmap='plasma', vmin=0, vmax=1)
plt.colorbar(label="Ground State Overlap")
plt.xlabel("$V/J$")
plt.ylabel("$T$")
plt.title(f"Ground State Overlap: $N={N}$, $M={M}$, $J_0 = {init_J}$, $V_0 = {init_V}$, $\\mu_0 = {init_mu}$, $J_f = 1$, $dt = {dt}$")
plt.show()
```
