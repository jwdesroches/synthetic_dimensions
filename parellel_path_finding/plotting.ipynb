```python
import sys
import os

parent_dir = os.path.abspath(os.path.join(os.getcwd(), os.pardir))
sys.path.append(parent_dir)

from synth_dim_model import *

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap
import pandas as pd

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("compiled_results.csv")
```

```python
M = 3
J_V_finals = [0, 0.5, 2]
Vs = [-1, 1]
colors = {2: "r", 3: "g", 4: "b", 5: "k"}

for V in Vs:
    for J_V_final in J_V_finals:
        fig, ax = plt.subplots()
        for N in [2, 3, 4, 5]:
            # Filter the data for this set of parameters
            subset = df[(df["N"] == N) & (df["J_V_final"] == J_V_final) & (df["V"] == V)]
            # For each time T, select the row with the maximum GROUND_STATE_OVERLAP (the "best" result)
            best_indices = subset.groupby("T")["GROUND_STATE_OVERLAP"].idxmax()
            d_best = subset.loc[best_indices].sort_values("T")
            
            ax.plot(d_best["T"], d_best["GROUND_STATE_OVERLAP"], '.', color=colors[N], label=f"$N={N}$")
            ax.plot(d_best["T"], d_best["GROUND_STATE_OVERLAP"], '--', color=colors[N])

        ax.set_ylim(-0.1, 1.1)
        ax.set_xlim(-0.1, 10.1)
        ax.grid()
        ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=4, fancybox=True, shadow=True)
        ax.set_title(f"Ground State Overlap: $M={M}$, $(J/|V|)_f = {J_V_final}$, $V = {V}$")
        ax.set_ylabel("Overlap")
        ax.set_xlabel("Total Path Time")
        fig.tight_layout()
```
